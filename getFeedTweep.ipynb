```python
import tweepy
```

```python
api_key =""
api_secret_key =""
access_token = "-"
access_token_secret = ""
```

```python
auth = tweepy.OAuthHandler(api_key,api_secret_key)
auth.set_access_token(access_token,access_token_secret)
api = tweepy.API(auth)
```

```python
crawlTweep = api.user_timeline(id="israel", count =10)
crawlTweep
```

```python
for tweet in crawlTweep:
  print(tweet.text)
```
