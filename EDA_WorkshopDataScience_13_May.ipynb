```python
df = df.replace('ERR', np.nan)
```

```python
df.head()
```

```python
# import library penanganan missing values
from pandas.core import missing
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
df[['value','item purchased','monthly visits']] = imputer.fit_transform(df[['value','item purchased','monthly visits']])
```

```python
# DROP
df.dropna(inplace=True)
df.head()
```

```python
#menghilangkan whitespce
df['gender'] = df['gender'].str.strip()
```

```python
df['country'] = df['country'].str.strip()
```

```python
df.info()
```

```python
# datetime
df['date'] = pd.to_datetime(df['date'] )
df.info()
```

```python
df.head()
```

```python
df['age'] = df['age'].astype(int)
```

```python
df.info()
```

```python
# penanganan anomaly data
df['country'].unique()
```

```python
df['country'] = df['country'].replace('USA', 'US')
```

```python
df['country'].unique()
```

```python
# mapping categorical data menjadi numerik
df['gender'] = df['gender'].map({'Male' :0, 'Female':1})
```

```python
df
```

```python
df.groupby('country').sum()
```

```python
def fungsiku(label):
    if label<500:
      return 'Normal'
    else:
      return 'Aktif'
```

```python
df['keterangan'] = df['monthly visits'].apply(lambda x : fungsiku(x))
```

```python
df.head()
```

```python
df['keteranga'] = df['keterangan'].map({'Normal' :0, 'Aktif':1})

df.head()
```

```python
df  = pd.get_dummies(df, columns=['country'])
```

```python
df.head()
```

```python
from scipy import stats

out_list = ['value','monthly visits','item purchased']

df = df[(np.abs(stats.zscore(df[out_list]))<3).all(axis=1)]

df.head()
```
